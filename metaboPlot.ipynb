```python
hover2 = HoverTool(tooltips=[
    ("Compound Name", "@Compound"),
    ("(x,y)", "($x, $y)"),
    ("Mass", '@Mass'),
    ("RT", '@RT')
])
tools = [hover2, BoxZoomTool(), ResetTool(), CrosshairTool(), BoxSelectTool(), WheelZoomTool(), SaveTool()]
tools
df = input1
df2=input2
```

```python
#Adjustable parameters
xis = "a"
yis = "d"
fc_value=0.05

#No need to touch from here on down
ycol='black'
if xis == "a":
    xnam="Unconcentrated"
elif xis == "b":
    xnam="Eppendorf"
elif xis == "c":
    xnam="Genevac"
else:
    xnam="Thermo"

if yis == "a":
    ynam="Unconcentrated"
elif yis == "b":
    ynam="Eppendorf"
    ycol='black'
elif yis == "c":
    ynam="Genevac"
    ycol='red'
else:
    ynam="Thermo"
    ycol='green'

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(df2[xis], df2[yis])

TITLE = "Ion Count Plot - "+str(xnam)+" vs. "+str(ynam)+"  (R-squared = "+str("%.4f" % r_value**2)+")"+"  (Fold-change = "+str(fc_value)+")"
TOOLS = "hover,pan,wheel_zoom,box_zoom,reset,save"

p = figure(tools=tools, toolbar_location="above", logo="grey", plot_width=900, 
          plot_height=900, title=TITLE, x_axis_type="log", y_axis_type="log",
          x_range=(100,1000000000), y_range=(100,1000000000))
p.background_fill_color = "#dddddd"
p.xaxis.axis_label = "Ion Counts, "+str(xnam)
p.yaxis.axis_label = "Ion Counts, "+str(ynam)
p.grid.grid_line_color = "white"


#source1 = ColumnDataSource(df)
source2 = ColumnDataSource(df2)
source3= ColumnDataSource(df2.loc[df2['fc a vs all']<fc_value])

p.ray(x=100, y=100, length=1000000000000, angle=0.2479*pi, color='black', line_dash="dashed")

p.circle(xis, yis, size=3, source=source2, color='tan')
p.square(xis, yis, size=6, source=source3, color=ycol)
show(p)
```
