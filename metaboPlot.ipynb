```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline
import holoviews as hv
import bokeh
hv.extension('bokeh')
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool, BoxZoomTool, CrosshairTool, ResetTool, BoxSelectTool, WheelZoomTool, SaveTool
from bokeh.models.renderers import GlyphRenderer
import copy
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, show, output_file
from math import pi
```

```python
excel_file = 'C:/Users/Davinder Sandhu/Desktop/180510_qtof_glyco_untargeted_merged.xlsx'
input1 = pd.read_excel(excel_file)#, index_col=0)
input1.head()
#input1['Compound Name']
df=input1
df.rename(columns={'Compound Name':'Compound'}, inplace=True)
```

```python
hover2 = HoverTool(tooltips=[
    ("Compound Name", '@Compound'),
    ("(x,y)", "($x, $y)"),
    ("Mass", '@Mass'),
    ("RT", '@RT')
])
tools = [hover2, BoxZoomTool(), ResetTool(), CrosshairTool(), BoxSelectTool(), WheelZoomTool(), SaveTool()]
tools
df = input1
```

```python
TITLE = "Ion Count Plot - Unconcentrated vs. Eppendorf"
TOOLS = "hover,pan,wheel_zoom,box_zoom,reset,save"

p = figure(tools=tools, toolbar_location="above", logo="grey", plot_width=900, 
          plot_height=900, title=TITLE, x_axis_type="log", y_axis_type="log",
          x_range=(100,1000000000), y_range=(100,1000000000))
p.background_fill_color = "#dddddd"
p.xaxis.axis_label = "Ion Counts, Unconcentrated"
#p.x_axis_type="log"
p.yaxis.axis_label = "Ion Counts, Eppendorf"
p.grid.grid_line_color = "white"

source = ColumnDataSource(df)
p.ray(x=100, y=100, length=1000000000000, angle=0.2479*pi, color='black', line_dash="dashed")
p.circle("b", "c", size=5, source=source)
show(p)
```
